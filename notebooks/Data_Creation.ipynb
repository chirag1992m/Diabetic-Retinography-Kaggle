```lua
require 'torch'
require 'torchnet'
require 'image'
Plot = require 'itorch.Plot'
```

```lua
file = io.open('../data/train_data/trainLabels.csv', "rb")

lines = {}
for line in file:lines() do
    lines[#lines + 1] = line
end
```

```lua
eye_train_data = {}
for i=2, #lines do
    track, eye, class_label = lines[i]:match("([^,]+)_([^,]+),([^,]+)")
    index = tonumber(track)
    class = tonumber(class_label)
    if eye == 'left' then
        eyetype = 1
    else
        eyetype = 2
    end
    if eye_train_data[index] == nil then
        eye_train_data[index] = {}
    end
    eye_train_data[index][eyetype] = class_label
end
```

```lua
length = 0
for k, v in pairs(eye_train_data) do
    length = length+1
end
```

```lua
-- Index 1 --File track
-- Index 2 --Left file available
-- Index 3 --Left eye label(if available)
-- Index 4 --Right eye available
-- Index 5 --Right eye label(if available)
eye_train_tensor = torch.LongTensor(length, 5)
```

```lua
index = 1
for k, v in pairs(eye_train_data) do
    eye_train_tensor[index][1] = k
    if v[1] == nil then
        eye_train_tensor[index][2] = -1
        eye_train_tensor[index][3] = -1
    else
        eye_train_tensor[index][2] = 1
        eye_train_tensor[index][3] = v[1]
    end
    
    if v[2] == nil then
        eye_train_tensor[index][4] = -1
        eye_train_tensor[index][5] = -1
    else
        eye_train_tensor[index][4] = 1
        eye_train_tensor[index][5] = v[2]
    end
    index = index + 1
end
```

```lua
torch.save('../data/train_full.t7', eye_train_tensor)
```

```lua
local dir = io.popen("ls ../data/sample/")

lines = {}
for line in dir:lines() do
    lines[#lines + 1] = line
end

eye_sample_data = {}
for i=1, #lines do
    track, eye, file_extension = lines[i]:match("([^_]+)_([^.]+).([^,]+)")
    index = tonumber(track)
    if eye == "left" then
        eyetype = 1
    else
        eyetype = 2
    end
    if eye_sample_data[index] == nil then
        eye_sample_data[index] = {}
    end
    eye_sample_data[index][eyetype] = 1
end
```

```lua
length = 0
for k, v in pairs(eye_sample_data) do
    length = length+1
end
```

```lua
-- index 1 --> the track of file
-- 2 --> Left is available
-- 3 --> Right is available
eye_sample_tensor = torch.LongTensor(length, 3)
```

```lua
index = 1
for k, v in pairs(eye_sample_data) do
    eye_sample_tensor[index][1] = k
    if v[1] == nil then
        eye_sample_tensor[index][2] = -1
    else
        eye_sample_tensor[index][2] = 1
    end
    
    if v[2] == nil then
        eye_sample_tensor[index][3] = -1
    else
        eye_sample_tensor[index][3] = 1
    end
    index = index + 1
end
```
