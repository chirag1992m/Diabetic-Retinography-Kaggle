```lua
require 'torch'
require 'torchnet'
require 'image'
Plot = require 'itorch.Plot'
```

```lua
file = io.open('../data/train_data/trainLabels.csv', "rb")

lines = {}
for line in file:lines() do
    lines[#lines + 1] = line
end
```

```lua
eye_train_data = {}
for i=2, #lines do
    track, eye, class_label = lines[i]:match("([^,]+)_([^,]+),([^,]+)")
    index = tonumber(track)
    class = tonumber(class_label)
    if eye == 'left' then
        eyetype = 1
    else
        eyetype = 2
    end
    if eye_train_data[index] == nil then
        eye_train_data[index] = {}
    end
    eye_train_data[index][eyetype] = class_label
end
```

```lua
file = io.open('../data/train_data/train_crop_data.txt', "rb")

lines = {}
for line in file:lines() do
    lines[#lines + 1] = line
end
```

```lua
eye_crop_data = {}
for i=1, #lines do
    track, eye, extension, y0, x0, y1, x1 = lines[i]:match("([^,]+)_([^.]+).([^,]+),([^,]+),([^,]+),([^,]+),([^,]+)")
    index = tonumber(track)
    y0, x0, y1, x1 = tonumber(y0), tonumber(x0), tonumber(y1), tonumber(x1)
    if eye == 'left' then
        eyetype = 1
    else
        eyetype = 2
    end
    if eye_crop_data[index] == nil then
        eye_crop_data[index] = {}
    end
    eye_crop_data[index][eyetype] = {x0, y0, x1, y1}
end
```

```lua
length = 0
for k, v in pairs(eye_train_data) do
    length = length+1
end
```

```lua
-- Index 1 --File track
-- Index 2 --Left or Right Eye
-- Index 3 --Pairing Eye Index
-- Index 4 --Cropping X0
-- Index 5 --Cropping Y0
-- Index 6 --Cropping X1
-- Index 7 --Cropping Y1
eye_train_tensor = torch.LongTensor(length*2, 7)
```

```lua
index = 1
for k, v in pairs(eye_train_data) do
    -- Left eye
    eye_train_tensor[index][1] = k
    eye_train_tensor[index][2] = 1
    eye_train_tensor[index][3] = index+1
    
    --Dont'have the cropping information yet
    eye_train_tensor[index][4] = eye_crop_data[k][1][1]
    eye_train_tensor[index][5] = eye_crop_data[k][1][2]
    eye_train_tensor[index][6] = eye_crop_data[k][1][3]
    eye_train_tensor[index][7] = eye_crop_data[k][1][4]
    index = index + 1
    
    -- Right eye
    eye_train_tensor[index][1] = k
    eye_train_tensor[index][2] = 2
    eye_train_tensor[index][3] = index - 1
    
    --Don't have the cropping information yet
    eye_train_tensor[index][4] = eye_crop_data[k][2][1]
    eye_train_tensor[index][5] = eye_crop_data[k][2][2]
    eye_train_tensor[index][6] = eye_crop_data[k][2][3]
    eye_train_tensor[index][7] = eye_crop_data[k][2][4]
    index = index + 1
end
```
